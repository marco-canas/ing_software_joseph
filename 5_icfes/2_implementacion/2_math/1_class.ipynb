## 📐 Matemáticas

### Categorías y Contenidos:

#### 1. Álgebra y Cálculo
- Expresiones algebraicas.
- Funciones (lineales, cuadráticas, exponenciales, logarítmicas).
- Sucesiones y límites.
- Porcentajes y razones de cambio.
- Notación científica.

#### 2. Geometría
- Figuras planas y sólidas (triángulos, círculos, cilindros, esferas).
- Perímetros, áreas y volúmenes.
- Coordenadas cartesianas.
- Transformaciones (traslación, rotación, simetría, etc.).
- Teoremas de Pitágoras y Tales.

#### 3. Estadística y Probabilidad
- Gráficas y tablas.
- Media, mediana, moda, rango.
- Varianza y percentiles.
- Combinaciones y permutaciones.
- Probabilidad básica e inferencia muestral.

---




### 🎯 Objetivo general
Evaluar la capacidad de resolver problemas mediante el uso de herramientas matemáticas. Se valora el razonamiento más allá del cálculo.

### 🧠 Competencias evaluadas
1. **Interpretación y representación (34%)**
   - Leer gráficos, tablas, esquemas y transformarlos.
   - Identificar patrones y relaciones.

2. **Formulación y ejecución (43%)**
   - Plantear estrategias de solución.
   - Usar operaciones y razonamientos lógicos.

3. **Argumentación (23%)**
   - Justificar respuestas o refutar soluciones.
   - Validar procesos matemáticos.
